# Pandas: grouping

```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("vehicles.csv")
```

```python
cars.head()
```

<font size="4"><font color='orange'>- How many Car models are there in the database? </font>

```python
## number of Models

len(cars["Model"].unique())
```

<font size="4"><font color='orange'>- Group by the data by the Make  using count function </font>

```python
## DF Grouped by "Make"

cars_grouped_by_Make = cars.groupby(["Make"]).count()
cars_grouped_by_Make
```

<font size="4"><font color='orange'>- What brand has the most cars? </font>

```python
## Sort counted values [=> chevrolet]  

cars_grouped_by_Make.sort_values(by="Model", ascending=False)
```

<font size="4"><font color='orange'>- Show the average CO2_Emission_Grams/Miles by Brand </font>

```python
## Groupby brand applied on COs emission

CO2_brands_avg = cars.groupby("Make")["CO2 Emission Grams/Mile"].mean()
CO2_brands_avg
```

<font size="4"><font color='orange'>- Show the average CO2_Emission_Grams/Miles by Brand ... sorted </font>

```python
CO2_brands_avg.sort_values(ascending=False)
```

<font size="4"><font color='orange'>- Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade. </font>

```python
cars['Year'].describe()
```

```python
## Defining bins according to decade 

cars["Decade"] = pd.cut(cars["Year"], bins=[1980, 1989, 1999, 2009, 2019], labels = ["80s", "90s", "2000s", "2010s"], retbins=False)
```

```python
cars
```

```python
## Grouping by decade + filtering out excess info 

cars.groupby("Decade")["Decade"].count()
```

<font size="4"><font color='orange'>- Did cars consume more gas in the eighties? </font>

show the average City_Km/Liter by year_range

```python
## Average gas consumption per decade

consumption_decade = cars.groupby(["Decade"])["City MPG"].mean()
consumption_decade
```

```python
consumption_decade.sort_values(ascending=True)
```

<font size="4"><font color='orange'>- Which brands are more environment friendly? </font>

```python
CO2_brands_avg.sort_values(ascending=True)
```

<font size="4"><font color='orange'>- Does the drivetrain affect fuel consumption? </font> 

```python
## Grouping fuel consumption by Drivetrain 

fuel_by_drivetrain = cars.groupby(["Drivetrain"])["Fuel Barrels/Year"].mean()
```

```python
fuel_by_drivetrain.sort_values(ascending=True)
```
